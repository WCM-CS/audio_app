```python
from yt_dlp
import ssl
import certifi
from urllib import request
from pytube import YouTube
# Note I had to update certifi package in py python version 3.11
#/Applications/Python\ 3.11/Install\ Certificates.command
```

```python
# Create an SSL context using the certifi certificate bundle
ssl_context = ssl.create_default_context(cafile=certifi.where())

# Ensure the SSL context is applied globally to urllib requests, used internally via YT download library
request.install_opener(request.build_opener(request.HTTPSHandler(context=ssl_context)))
```

```python
# URL's
url_list = [
    "https://www.youtube.com/watch?v=tWAqijudEqU", # Pure Noise
    "https://www.youtube.com/watch?v=CefKtc2YOak", # Gym Phonk
    "https://www.youtube.com/watch?v=n61ULEU7CO0", # Lofi Long
    "https://www.youtube.com/watch?v=iicfmXFALM8", # Lofi Short
    "https://www.youtube.com/watch?v=bla77zK9aM8", # Top Classical
    "https://www.youtube.com/watch?v=26Ch2vKsgTg&t=615s" # Dark Classical 
]
```

```python
# Download block

# Define format & specs
ydl_opts = {
        'format': 'best',
        'postprocessors': [{
            'key': 'FFmpegExtractAudio',
            'preferredcodec': 'mp3',  
            'preferredquality': '192',  
        }],
        'outtmpl': '%(title)s.%(ext)s',  # Save with title as filename
    }

# Run loop downloading the url's
for url in url_list:
    try:
        print(f"Starting Download for: {url}")
        with yt_dlp.YoutubeDL(ydl_opts) as ydl:
            ydl.download([url])
        print(f"Download for {url} is Complete.")
    except Exception as e:
        print(f"Error while downloading {url}: {e}")
```
